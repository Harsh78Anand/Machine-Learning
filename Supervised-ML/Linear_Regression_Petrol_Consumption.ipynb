# Petrol Consumption

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

i) *Histogram*

```
sample_data=pd.read_csv('/content/sample_data/petrol_consumption.csv')
df = pd.DataFrame(sample_data)
df.hist(figsize=(10,10))
plt.show()
```

ii) *Density Plot*

```
df.plot(kind='density', layout=(3,3), subplots=True, figsize=(10,10))
plt.show()
```

iii) *Boxplot*

```
df.plot(kind='box', layout=(3,3), subplots=True, figsize=(10,10))
plt.show()
```

**Normalisation**

```
from IPython.display import display
from sklearn.preprocessing import Normalizer
normalizer = Normalizer()
normal_data = normalizer.fit_transform(df)
print(normal_data)
```

**Scaling**

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
print(scaled_data)
```

**Standardisation**

```
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
standard_data = standard.fit_transform(df)
print(standard_data)
```

**Binarizer**

```
from sklearn.preprocessing import Binarizer
binarizer = Binarizer(threshold=7)
bin_data = binarizer.fit_transform(df)
print(bin_data)
```

**Multilinear Regression**

```
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
import seaborn as sns
corr=df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
plt.show()
```

```
X=sample_data.iloc[:,:4]
Y=sample_data.iloc[:,4:]
sns.distplot(Y,bins=50)
plt.show()
```

```
from sklearn.model_selection import train_test_split
X_nor=normalizer.fit_transform(X)
X_train,X_test,Y_train,Y_test=train_test_split(X_nor,Y,test_size=.50,random_state=0)
lr_model=LinearRegression()
lr_model.fit(X_train,Y_train)
Y_predict=lr_model.predict(X_test)
from sklearn.metrics import accuracy_score
print("Training Score",lr_model.score(X_train,Y_train))
print("Test Score",lr_model.score(X_test,Y_test))
print('R2 Score ',r2_score(Y_test,Y_predict))
```
